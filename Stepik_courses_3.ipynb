```python
%matplotlib inline
import statistics
from random import randint
import random as r
import math 
import numpy as np
from numpy import random
from scipy import (misc, stats, ndimage)
import pandas as pd 
import statsmodels.api as sm
from statsmodels.formula.api import ols 
from statsmodels.stats.multicomp import (pairwise_tukeyhsd, MultiComparison)
from statsmodels.graphics.factorplots import interaction_plot
import matplotlib.pyplot as plt 
import seaborn as sns 
from mpl_toolkits.mplot3d import Axes3D
```

```python
'''Демонстрация ковариации и корреляции'''
def covariation(x, y):
    assert x.size == y.size
    return ((x - x.mean() * (y - y.mean())).sum() / (x.size - 1))

def correlation(x, y):
    return covariation(x, y) / (np.std(x, ddof=1) * np.std(y, ddof=1))

def randomize(arr, p):
    alpha = np.max(arr) - np.min(arr)
    res  = np.zeros(arr.shape)
    for i, j in enumerate(arr):
        sign = 1 if r.choice([True, False]) else -1
        res[i] = j + sign * alpha * r.random() * p
    return res
```

```python
x = np.array(range(50))
y = randomize(x, 0.1)
y1 = randomize(x, 0.5)
y2 = randomize(x, 1)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(17, 3))
ax1.scatter(x, y)
ax2.scatter(x, y1)
ax3.scatter(x, y2)
ax1.set_title("Высокая корреляция")
ax2.set_title("Средняя корреляция")
ax3.set_title("Низкая корреляция")
plt.show()
```
